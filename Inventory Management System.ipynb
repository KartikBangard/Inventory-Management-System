```python
import json

Sale = ' '

# Importing Inventory data from Record.json file
fd = open("Record.json","r")
js = fd.read()
fd.close()

# Converting String data to Dictionary
record = json.loads(js)

# Displaying Menu
print("---------------MENU-----------------\n")

for key in record.keys():
    print(key,':',record[key]['Name'],'|',record[key]['Price'],'|',record[key]['Qty'])
    
print("------------------------------------\n")

# Taking Inputs from the user about their details and purchase
ui_name       = str(input("Enter Customer Name  :"))
ui_ph         = str(input("Enter Phone No.      :"))
ui_email      = str(input("Enter email_id       :"))
ui_product_id = str(input("Enter Product ID     :"))
ui_qty        = int(input("Enter Quantity       :"))


# If we're having more than or equal to the quantity user wants
if (record[ui_product_id]['Qty']) >= ui_qty:

    print("------------------------------------")
    print("               Bill                 \n")
    print("Product Name   - " , record[ui_product_id]['Name'])
    print("Price          - ",record[ui_product_id]['Price'])
    print("Quantity       - ", ui_qty)
    print("------------------------------------")
    print("Total Payable  - ",record[ui_product_id]['Price']*ui_qty)
    print("------------------------------------")

    # Updating Inventory in Dictionary
    record[ui_product_id]['Qty'] = record[ui_product_id]['Qty'] - ui_qty 

    
    # Generating CSV Transaction Detail
    Sale = ui_name+","+ui_ph+","+ui_email+","+ui_product_id+","+record[ui_product_id]["Name"]+","+str(ui_qty)+","+str(record[ui_product_id]["Price"])+","+str(ui_qty*record[ui_product_id]["Price"])+'\n'

    # If we have less quantity than the user wants
else:
    print("Product is about to be out of Stock")
    print("Only ",record[ui_product_id]['Qty'],"units are left")
    print("------------------------------------")
    ch=str(input("Press Y/y to place a order: "))
    
    # If user wants to purchase the whole quantity for that product
    if (ch == "Y" or ch == "y"):
           
        print("------------------------------------")
        print("               Bill                 \n")
        print("Product Name   - " , record[ui_product_id]['Name'])
        print("Price          - ",record[ui_product_id]['Price'])
        print("Quantity       - ", record[ui_product_id]['Qty'])
        print("------------------------------------")
        print("Total Payable  - ",record[ui_product_id]['Price']*record[ui_product_id]['Qty'])
        print("------------------------------------")
        print("------------------------------------")
        
        # Generating CSV Transaction Detail
        Sale = ui_name+","+ui_ph+","+ui_email+","+ui_product_id+","+record[ui_product_id]["Name"]+","+str(record[ui_product_id]["Qty"])+","+str(record[ui_product_id]["Price"])+","+str(record[ui_product_id]["Qty"]*record[ui_product_id]["Price"])+'\n'
        
        # Updating Inventory in Dictionary
        record[ui_product_id]["Qty"] = 0
        
      # If user gave any input other than Y or y
    else:
         print("Thanks for visiting!")

# Converting Inventory Dictionary to String
js = json.dumps(record)

# Updating Inventory and Saving in to my Records.json
fd = open("Sales.txt",'a')
fd.write(Sale)
fd.close()

# Adding Transaction on Sales File
fd = open("Record.json",'w')
fd.write(js)
fd.close()
```
